```python
from diffusers import StableDiffusionPipeline
from matplotlib import pyplot as plt
import torch
import os


model_id = "runwayml/stable-diffusion-v1-5"
meina = r"C:\Users\fired\.cache\huggingface\hub\models--runwayml--stable-diffusion-v1-5\snapshots\451f4fe16113bff5a5d2269ed5ad43b0592e9a14\unet\meinapastel_v6Pastel32.safetensors"
anima_opt = r"E:\D1\diffusion\stable-diffusion-webui-1.10.1\models\Stable-diffusion\v1-5-pruned-emaonly.safetensors"
stable_diffusion_txt2img = StableDiffusionPipeline.from_single_file(anima_opt, torch_dtype=torch.float16)
stable_diffusion_txt2img.to("cuda")
```
